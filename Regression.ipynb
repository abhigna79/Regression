6. What is the purpose of the least squares method in Simple Linear Regression?
   - To find the best-fitting line through the data by minimizing the sum of the squared differences between the actual data points and the predicted values from the line.

7.  How is the coefficient of determination (R¬≤) interpreted in Simple Linear Regression?

    - R¬≤ measures how well the regression line fits the data.
    - it tells you how much of the variation in the dependent variable (Y) is explained by the independent variable (X) in your model.

9. What is the main difference between Simple and Multiple Linear Regression?
   - Simple linear regression:
   - Only one predictor (X)

11.  What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?
    - What is Heteroscedasticity?
    - Heteroscedasticity refers to a situation where the variance of the residuals (errors) in a regression model is not constant across all levels of the independent variables.

    - How Does Heteroscedasticity Affect a Multiple Linear Regression Model?
    - Bias in Coefficient Estimates
    - Incorrect Inferences
    - Confidence Intervals
    - Inefficient Predictions

12. How can you improve a Multiple Linear Regression model with high multicollinearity?

31. How is polynomial regression implemented in Python?
    - To implement polynomial regression in Python, we can use scikit-learn and matplotlib libraries for model fitting and visualization, respectively.
    - Import Libraries
    - Generate or Load Data
    -  Visualize the Data
    - Polynomial Feature Transformation
    - Fit the Polynomial Regression Model
    - Make Predictions
    - Visualize the Polynomial Regression Fit
    - EXAMPLE

   





   












```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 1: Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Random X values between 0 and 10
y = 3 * X**2 + 2 * X + 5 + np.random.randn(100, 1) * 10  # Quadratic relationship with noise

# Step 2: Visualize the data
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of Data')
plt.show()

# Step 3: Polynomial Feature Transformation
degree = 2  # Degree of the polynomial (e.g., quadratic)
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# Step 4: Fit Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, y)

# Step 5: Make Predictions
y_pred = model.predict(X_poly)

# Step 6: Visualize the Polynomial Fit
X_grid = np.arange(min(X), max(X), 0.1)  # Smooth curve
X_grid = X_grid.reshape((len(X_grid), 1))
X_grid_poly = poly.transform(X_grid)
y_grid_pred = model.predict(X_grid_poly)

# Plotting
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_grid, y_grid_pred, color='red', label=f'Polynomial fit (degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
